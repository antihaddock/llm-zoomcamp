```python
# #Question 6
import openai

# Configure OpenAI client for Ollama
openai.api_base = "http://localhost:11434/v1/"  # Ollama default address
openai.api_key = "ollama"  # Not actually used, but required by the library

# Your prompt
prompt = "What's the formula for energy?"

client = openai.OpenAI(api_key="ollama", base_url="http://localhost:11434/v1")

# Create chat completion request with temperature 0
response = client.chat.completions.create(
    model="gemma:2b",
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": prompt},
    ],
    temperature=0
)

# Get the model's response
answer = response.choices[0].message.content

# Get the number of completion tokens
completion_tokens = response.usage.completion_tokens

print(f"Answer: {answer}")
print(f"Completion Tokens Used: {completion_tokens}")
```
